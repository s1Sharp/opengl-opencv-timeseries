```python
import numpy as np
import matplotlib.pyplot as plt

#    
def one_experiment(steps=10000):
    happiness = [0]  #   0  ""
    for _ in range(steps):
        step = np.random.choice([-1, 1], size=1)[0]  #     ""  1
        happiness.append(int(happiness[-1]) + step)
    answ = np.array(happiness)
    answ = answ[answ != 0]
    return answ > 0  #   : True,  "" > 0,  False

#  100 
num_experiments = 1000
experiment_results = []
for exp in range(num_experiments):
    experiment_results.append(one_experiment())

#   ""    
happiness_ratios = [np.mean(experiment) for experiment in experiment_results]
```

```python
#    "" 
plt.figure(figsize=(30, 5))
plt.plot(np.arange(1, num_experiments + 1), happiness_ratios, marker='o', linestyle='-', color='b')
plt.axhline(0.5, color='r', linestyle='--', label="  (0.5)")
plt.xlabel(" ")
plt.ylabel(" '' ")
plt.legend()
plt.title("  '' ")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(happiness_ratios, bins=100, density=True, color='blue', alpha=0.7)
plt.axvline(0.5, color='red', linestyle='--', label=" .  (0.5)")
plt.xlabel("  ''")
plt.ylabel("")
plt.legend()
plt.title("   ''")
plt.grid()
plt.show()
```
