```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для одного эксперимента
def one_experiment(steps=10000):
    happiness = [0]  # Начинаем с 0 уровня "счастья"
    for _ in range(steps):
        step = np.random.choice([-1, 1], size=1)[0]  # Случайно увеличиваем или уменьшаем "счастье" на 1
        happiness.append(int(happiness[-1]) + step)
    answ = np.array(happiness)
    answ = answ[answ != 0]
    return answ > 0  # Возвращаем булевский массив: True, если "счастье" > 0, иначе False

# Проводим 100 экспериментов
num_experiments = 1000
experiment_results = []
for exp in range(num_experiments):
    experiment_results.append(one_experiment())

# Рассчитываем доли "счастливых" моментов в каждом эксперименте
happiness_ratios = [np.mean(experiment) for experiment in experiment_results]
```

```python
# Построение графика доли "счастливых" моментов
plt.figure(figsize=(30, 5))
plt.plot(np.arange(1, num_experiments + 1), happiness_ratios, marker='o', linestyle='-', color='b')
plt.axhline(0.5, color='r', linestyle='--', label="Теоретическая доля (0.5)")
plt.xlabel("Номер эксперимента")
plt.ylabel("Доля 'Счастливых' Моментов")
plt.legend()
plt.title("Экспериментальная доля 'Счастливых' Моментов")
plt.grid()
plt.show()
```

> тогда рассмотрим формулу: 
* Случайная величина X на [0,1] распределяется по арксину, если что. Пусть



```python
P_1_4 = 2/np.pi * np.arcsin(np.sqrt(1/4))
P_1_4
```

```python
P_2_4 = 2/np.pi * np.arcsin(np.sqrt(2/4)) - P_1_4
P_2_4
```

```python
P_3_4 = 2/np.pi * np.arcsin(np.sqrt(3/4)) - P_1_4 - P_2_4
P_3_4
```

```python
P_4_4 = 2/np.pi * np.arcsin(np.sqrt(4/4)) - P_1_4 - P_2_4 - P_3_4
P_4_4
```
