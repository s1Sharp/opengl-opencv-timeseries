```python
import cv2

# BGR
img = cv2.imread('img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_rgb)
```

```python
_, ax = plt.subplots(1, 2, figsize=(20,10))
ax[0].imshow(img)
ax[0].set_title('bgr')
ax[1].imshow(img_rgb)
ax[1].set_title('rgb')
plt.show()
```

```python
cv2.imwrite('img_bgr.png', img_rgb)
```

```python
Image.fromarray(img_rgb).save('img_pil.png')
```

```python
img_gray = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, 'gray')
```

```python
_, img_binary = cv2.threshold(img_gray, 90, 255, cv2.THRESH_BINARY)
plt.imshow(img_binary, 'gray')
```
